```python
## Get date column in the right format
```

```python
import datetime
```

```python
sales['Date'] = pd.to_datetime(sales.Date, format='%d/%m/%Y')
```

```python
sales['Day'] = sales.Date.dt.day
```

```python
sales['Month'] = sales.Date.dt.month
```

```python
sales['Year'] = sales.Date.dt.year
```

```python
sales.drop('Date',axis=1,inplace=True)
```

```python
sales
```

```python
## Sales by year
```

```python
v = sales.groupby('Year').sum()
plt.figure(figsize=(20,6))
sns.barplot(x=v.index,y=v.Weekly_Sales)
```

```python
for i in sales.Year.unique():
    k = sales.loc[sales.Year == i]
    v = k.groupby('Month').sum()
    plt.figure(figsize=(20,6))
    sns.barplot(x=v.index,y=v.Weekly_Sales)
    plt.title('Monthly Sales in ' + str(i))
```

Despite the limitations of the data, we can see that sales tend to increase during months 3-4 and then again on months 6-7 and also towards the end of the year.

```python
for i in sales.Year.unique():
    k = sales.loc[sales.Year == i]
    v = k.groupby('Day').mean()
    plt.figure(figsize=(20,6))
    sns.barplot(x=v.index,y=v.Weekly_Sales)
    plt.title('Daily Average Sales in ' + str(i))
```

```python
## by store
```

```python
for i in sales.Year.unique():
    k = sales.loc[sales.Year == i]
    v = k.groupby('Store').sum().sort_values('Weekly_Sales',ascending=False).head(10)
    plt.figure(figsize=(20,6))
    sns.barplot(x=v.index, y=v.Weekly_Sales, data=v)
    plt.title('Top 10 selling stores in ' + str(i))
```

Our graphs indicate that the stores which consistently have the highest yearly sales are 2, 4, 10, 13, and 20.

```python
## which depts sell the most from each of these stores?
```

```python
strs = [2,4,10,13,20]
years = [2010,2011,2012]

for i in years[0:]:
    z = sales.loc[sales.Year == i]
    for j in strs[0:]:
        v = z.loc[z.Store == j]
        k = v.groupby('Dept').sum()
        plt.figure(figsize=(20,5))
        sns.barplot(x=k.index,y=k.Weekly_Sales,data=k)
        plt.xlabel('Departments')
        plt.ylabel('Sales')
        plt.title(str(i) + ' Sales by Dept in Store ' + str(j))
```

For each of the stores we found the following information: 
- For store 2 dpts 90-95 and 38 and 40 are the ones with the highest sales in all three years
- For store 4 depts 38,40,72, and 90=95 have the highest sales
- For store 10 depts 2,8, 38, 40, and 72
- For store 13 depts 38, 40, and 90-95 
- For store 20 depts 38, 40, 92 and 95

```python
stores_full = pd.merge(sales, stores, on='Store', how='inner')
stores_full
```

```python
## check size in relation to sales, check type of store in relation to sales, type of store and number of dpts?
```

```python
stores_full['Sales_Size_Ratio'] = stores_full['Weekly_Sales']/stores_full['Size']
```

```python
stores_full.head()
```

```python
z = stores_full.groupby('Store').sum().sort_values('Sales_Size_Ratio',ascending=False).head(10)
```

```python
plt.figure(figsize=(20,6))
sns.barplot(data=z,x=z.index,y=z['Sales_Size_Ratio'])
plt.ylabel('Sales')
plt.title('Stores with the highest Sales to Size Ratio')
```

```python
s = stores_full.groupby('Store').mean().sort_values('Size',ascending=False).head(10)
```

```python
plt.figure(figsize=(20,6))
sns.barplot(data=s,x=s.index,y=s['Size'])
plt.title('Ten biggest stores by size')
```

```python
s_rev = stores_full.groupby('Store').mean().sort_values('Size',ascending=True).head(10)
```

```python
plt.figure(figsize=(20,6))
sns.barplot(data=s_rev,x=s_rev.index,y=s_rev['Size'])
plt.title('Ten Smallest Stores')
```

2 of the 10 biggest stores and 4 out of the smallest stores have some of the highest sales to size ratio. This indicates that there may not be a direct correlation between size and sales. However, we can do further analysis and determine the correlation with more accuracy.

```python
## Does having more departments play a role in sales?
```

```python
for i in stores_full.columns[0:2]:
    stores_full[i] = stores_full[i].apply(lambda x: str(x))
```

```python
for i in stores_full.Store.unique()[0:]:
    v = stores_full.loc[stores_full.Store == i]
    print('Store ' + i + ' has ' + str(len(v.Dept.unique())) + ' Departments')
```

Since almost all stores have about the same number of departments it is unlikely that the number of departments each store has may influence the amount of sales in a significant way.
